## Image Pyramids

Useful when scaling images in object detection.

Many times, we have a need to change the size of the image for image analysis. For example, in machine learning applications, we may need to change the input size up or down so that we can train the network on different sized images. This can be achieved using cv2.pyrUp() and cv2.pyrDown(). This creates an "image pyramid".



Like this, the image size is halved or doubled at each iteration.

cv2.pyrUp(img) - This doubles the image dimensions.
cv2.pyrDown(img) - This halves the image dimensions.


```python
import cv2

image = cv2.imread('images/input.jpg')

smaller = cv2.pyrDown(image)
larger = cv2.pyrUp(smaller)

cv2.imshow('Original', image )

cv2.imshow('Smaller ', smaller )
cv2.imshow('Larger ', larger )
cv2.waitKey(0)
cv2.destroyAllWindows()
```
