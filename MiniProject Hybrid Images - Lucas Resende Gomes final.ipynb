## Blur the image

Coded by Lucas Resende Gomes

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread(r"C:\Users\lucas\OneDrive\Documentos\computer_vision\dataHybrid\dog.bmp")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kn1 = 2
blur = cv2.GaussianBlur(img,(25,25), kn1)


plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blured')
plt.xticks([]), plt.yticks([])
                                           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Sharpen the image

```python
import cv2
import numpy as np

image = cv2.imread(r"C:\Users\lucas\OneDrive\Documentos\computer_vision\dataHybrid\cat.bmp")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
kn2 = 10

sharpened = ((image) - ((cv2.GaussianBlur(image, (25,25), kn2))))-110

plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(sharpened),plt.title('Blured')
plt.xticks([]), plt.yticks([])
                                           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Main code

```python
hybrid = cv2.addWeighted(blur, 0.5, sharpened, 0.5, -1)
cv2.imshow('hello word', hybrid)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
