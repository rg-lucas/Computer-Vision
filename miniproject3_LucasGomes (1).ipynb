## Blur the image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread(r"C:\Users\lucas\OneDrive\Documentos\computer_vision\dataHybrid\dog.bmp")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kn1 = 5
blur = cv2.GaussianBlur(img,(25,25), kn1) -1

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blured')
plt.xticks([]), plt.yticks([])
                                           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Sharpen the image

```python
import cv2
import numpy as np

image = cv2.imread(r"C:\Users\lucas\OneDrive\Documentos\computer_vision\dataHybrid\cat.bmp")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
kn2 = 15

sharpened = image - cv2.GaussianBlur(image, (25,25), kn2)-120

plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(sharpened),plt.title('Blured')
plt.xticks([]), plt.yticks([])
                                           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Main code

```python
scale_percent = 60
width = int(img.shape[1] * scale_percent / 50)
height = int(img.shape[0] * scale_percent / 50)
dim = (width, height)
resized_blur = cv2.resize(blur, dim, interpolation= cv2.INTER_AREA)
resized_shapened = cv2.resize(sharpened, dim, interpolation = cv2.INTER_AREA)

hybrid = cv2.bitwise_and(resized_blur, resized_shapened)
hybrid = cv2.cvtColor(hybrid, cv2.COLOR_BGR2RGB)
cv2.imshow('hello word', hybrid)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
